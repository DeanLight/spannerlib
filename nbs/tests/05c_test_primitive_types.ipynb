```python
from spanner_workbench.rgxlog.primitive_types import DataTypes, DataTypeMapping, Span
import pytest
```

```python
def test_primitive_types_sanity() -> None:
    # test creating illegal Span instances:
    with pytest.raises(TypeError) as e_info:
        Span('x',0)
    assert str(e_info.value) == "Span's start/end must be integers" 

    with pytest.raises(TypeError) as e_info:
        Span('x','x')
    assert str(e_info.value) == "Span's start/end must be integers"

    with pytest.raises(TypeError) as e_info:
        Span(0,'x')
    assert str(e_info.value) == "Span's start/end must be integers" 

    span1 = Span(15,16)
    span2 = Span(15,16)
    assert span1 == span2

    span1 = Span(15,16)
    span2 = Span(15,17)
    assert span1 < span2

    span1 = Span(14,16)
    span2 = Span(13,17)
    assert span1 > span2

test_primitive_types_sanity()
```
